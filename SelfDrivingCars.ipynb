# 1.0 Introduction to Self Driving Cars

## 1.1 Taxonomy of Driving in Self-Driving Cars

1. **Driving Task**:
   - **Perceiving the Environment**: Self-driving cars utilize sensors (such as cameras, lidar, radar) to gather data about their surroundings, including other vehicles, pedestrians, road signs, and traffic conditions.
   - **Planning**: This involves determining the best route from point A to point B, considering factors like traffic, road conditions, and user preferences.
   - **Controlling the Vehicle**: The vehicle's control system manages actions like steering, braking, and accelerating, translating the planned trajectory into real-time movements.

2. **Operational Design Domain (ODD)**:
   - This defines where and under what conditions a self-driving car can operate safely. ODD outlines the geographical area, speed limits, weather conditions, and other constraints within which the vehicle can function autonomously.

3. **Lateral Control - Steering**:
   - Lateral control focuses on maintaining the car's position within a lane or executing safe lane changes. It involves controlling the steering to follow the desired path.

4. **Longitudinal Control - Braking, Accelerating**:
   - Longitudinal control manages the vehicle's speed by adjusting acceleration and braking. It ensures safe following distances and controlled deceleration when needed.

5. **Object and Event Detection and Response (OEDR)**:
   - **Detection**: Sensors identify objects, obstacles, pedestrians, and other vehicles in the environment.
   - **Reaction**: Based on detected objects and events, the self-driving car's system determines appropriate responses, such as slowing down, changing lanes, or stopping.

6. **Planning - Long Term (Route) - Short Term**:
   - **Long-Term Planning (Route)**: Involves selecting an optimal route from start to destination, considering factors like traffic patterns, road conditions, and user preferences.
   - **Short-Term Planning**: This involves making real-time decisions within the current driving context, including adjusting speed, choosing lanes, and responding to immediate changes in the environment.

7. **Miscellaneous**:
   - This category encompasses various additional elements, including regulatory compliance, user interfaces, communication with other vehicles and infrastructure, and redundancy systems to ensure safety.

## 1.3 Requirements for Perception

**Perception** is a critical aspect of autonomous driving that enables vehicles to understand and interact with their surroundings.

- **Identification and Understanding**: Perception entails identifying and comprehending various elements within the environment, including both static and dynamic objects.

- **Static Objects**: Autonomous vehicles must recognize and interpret various static objects on the road, such as road and lane markings, curbs, traffic lights, road signs, construction signs, obstructions, and more.

- **Dynamic Objects**: Recognizing dynamic objects is equally vital, encompassing the detection and tracking of moving entities such as vehicles, pedestrians, and animals.



**Goals for Perception**:

- **Ego Localization**: Determining the vehicle's own position accurately in the environment.

- **Position, Velocity, and Acceleration**: Understanding the position, speed, and rate of change of speed of surrounding objects.

- **Orientation and Angular Motion**: Recognizing the orientation and angular movements of objects, including their trajectories.


**Challenges to Perception**:
- **Robust Detection and Segmentation**: Accurately detecting and distinguishing objects from their surroundings, even in complex scenarios.

- **Sensor Uncertainty**: Dealing with the inherent uncertainty in sensor measurements and data.

- **Occlusion and Reflection**: Addressing instances where objects are hidden from view by other objects or reflections.

- **Illumination and Lens Flare**: Handling changes in lighting conditions and potential glare that could affect sensor readings.

- **Weather and Precipitation**: Adapting perception to different weather conditions, including rain, snow, fog, and glare.


## 1.4 Sensors in Self-Driving Cars

Sensors can be categorized into two main types: **exteroceptive** and **proprioceptive** sensors.

**Exteroceptive Sensors**:

These sensors focus on measuring properties of the external environment,

- **Cameras**: Mimicking human vision, cameras capture visual data, helping the car identify road signs, traffic lights, pedestrians, and other vehicles.
- **Lidar (Light Detection and Ranging)**: Lidar employs laser beams to measure distances and create detailed 3D maps of the surroundings, aiding in object detection and environment mapping.
- **Radar (Radio Detection and Ranging)**: Radar uses radio waves to detect objects and gauge their speed, contributing to collision avoidance and adaptive cruise control.
- **Ultrasonic Sensors**: These sensors use sound waves to measure distances and assist in parking and low-speed maneuvers.

**Proprioceptive Sensors**:

Proprioceptive sensors focus on monitoring the internal state of the vehicle, contributing to its stability and control.

- **IMU (Inertial Measurement Unit)**: IMUs combine accelerometers and gyroscopes to measure the vehicle's acceleration, orientation, and angular velocity.
- **Wheel Encoders**: Wheel encoders track the rotation of wheels, aiding in distance and speed calculations.
- **Steering Angle Sensor**: This sensor monitors the steering wheel's position, vital for accurate control and autonomous navigation.


**Perception Sensors**

In the realm of autonomous driving, perception sensors take center stage as the vigilant eyes and ears of self-driving vehicles. These sensors are pivotal in accurately comprehending the surrounding environment, making informed decisions, and ensuring passenger safety. When evaluating perception sensors, several crucial metrics come into play, including resolution, field of view, and dynamic range.

**Examples of Perception Sensors**:

- **Lidar (Light Detection and Ranging)**
- **Cameras**
- **Radar (Radio Detection and Ranging)**
- **Ultrasonic Sensors**


**Enhancing Perception**

**Robust Object Detection and Relative Speed Estimation**:

Robust object detection and accurate estimation of relative speeds are essential for ensuring the safety and efficiency of autonomous vehicles. These capabilities enable the vehicle to identify other vehicles, pedestrians, and obstacles in its path and gauge their movement to make precise driving decisions.

**Metrics for Comparison**:

1. **Range**

2. **Field of View**

**Configurations**:

1. **WFOV (Wide Field of View), Short Range**: Sensors with a wide field of view but limited range are suited for detecting objects in close proximity to the vehicle. They excel in urban environments and low-speed situations where detailed awareness of nearby obstacles is critical.

2. **NFOV (Narrow Field of View), Long Range**: Sensors with a narrower field of view and extended range are designed to identify objects at greater distances. These sensors are effective for highway driving and scenarios where perceiving distant vehicles is paramount.


**Global Navigation Satellite Systems (GNSS) and Inertial Measurement Units (IMU): Navigating Autonomy with Precision**

Global Navigation Satellite Systems (GNSS) and Inertial Measurement Units (IMUs) are foundational technologies that provide crucial data for autonomous vehicles, offering direct measurements of ego vehicle states. These technologies empower self-driving cars to accurately perceive their surroundings and make informed decisions. Key attributes and measurements associated with GNSS and IMUs are instrumental in enhancing the vehicle's understanding of its environment.

**GNSS - Direct Measure of Ego Vehicle States**:

GNSS systems utilize signals from a constellation of satellites to determine the vehicle's position, velocity, and heading. This direct measurement of ego vehicle states forms the basis of self-driving vehicle navigation.

- **Position and Velocity (GNSS)**: GNSS provides real-time position and velocity data. Various accuracy levels exist, including:
  - **Real-Time Kinematic (RTK)**: High-precision positioning achieved by using correction signals from reference stations.
  - **Precise Point Positioning (PPP)**: Achieves centimeter-level accuracy without the need for nearby reference stations.
  - **Differential GPS (DGPS)**: Corrects inaccuracies in standard GPS measurements, enhancing accuracy for navigation.

**IMU - Direct Measure of Ego Vehicle States**:

Inertial Measurement Units (IMUs) consist of accelerometers and gyroscopes to measure various vehicle motion parameters.

- **Angular Rotation Rate (IMU)**: IMUs measure the vehicle's angular rotation rate, allowing for accurate assessment of its orientation and trajectory.
- **Acceleration (IMU)**: Accelerometers in IMUs gauge the vehicle's acceleration, helping to determine changes in speed and direction.
- **Heading (IMU, GPS)**: IMUs can contribute to determining vehicle heading, enhancing understanding of the vehicle's orientation. Some GPS systems also provide heading information.
